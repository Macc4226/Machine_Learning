# Simple Linear Regression
1. Profit (Target) ~ RND (Predictor)
2. Profit_pred = B0 + B1 * RND
3. B0 = Intercept
4. B1 = Coefficient (Slope)

## Seperate X and Y

```python
X = df[["RND"]]
Y = df[["PROFIT"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(X, Y)
```

```python
model1.intercept_
```

```python
model1.coef_
```

### Profit_pred = 49032.90 + 0.854 x RND

## Evaluate the model

```python
# R2 Score on Data
model1.score(X, Y)
```

```python
sns.regplot(data=df, x="RND", y="PROFIT", line_kws={"color": "orange"})
```

```python
# Prediction given by straight line
preds = model1.predict(X)
preds[0:5]
```

```python
Y.head()
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score,
)


def evaluate_model(model, x, y):
    # Predict result for x
    ypred = model.predict(x)
    # Calculate the metrics
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print above results
    print(f"RMSE  : {rmse:.2f}")
    print(f"MAE   : {mae:.2f}")
    print(f"MAPE  : {mape:.2%}")
    print(f"R2    : {r2:.2%}")
```

```python
evaluate_model(model1, X, Y)
```

#### **ABOVE MODEL CLEARLY SHOWS THAT MODEL IS GOOD AND CAN BE USED FOR OUT OF SAMPLE PREDICTIONS**

### Multiple Linear Regression
1. Single Target ~ Multiple Predictors (Profit ~ RND + Admin + MKT)
2. Profit = B0 + B1 * RND + B2 * Admin + B3 * MKT



```python
# Multiple Predictor
X = df[["RND", "ADMIN", "MKT"]]
Y = df["PROFIT"]
```

```python
X.head()
```

```python
model2 = LinearRegression()
model2.fit(X, Y)
```

```python
model2.intercept_
```

```python
model2.coef_
```

```python
X.columns
```

```python
coefs = pd.Series(model2.coef_, index=X.columns)
coefs
```

### Profit_pred = 50122.1928994083 + 0.805715 * RND + -0.026816 * Admin + 0.027228 * MKT

```python
coefs.plot(kind="bar")
```

```python
# Evaluate the multiple regression model2
evaluate_model(model2, X, Y)
```

### R2 score > 0.8 hence model is good and can be used for out of sample predictions

```python
rnd = 1_20_000
mkt = 30_000
admin = 50_000

profit_pred = model2.predict([[rnd, admin, mkt]])
print(profit_pred)
```
